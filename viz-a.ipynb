```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

import data data from csv file and show it

```python
df_house = pd.read_csv('./kc_house_data.csv')
df_house
```

calculate avg price per room num

```python
average_price_per_room_num=df_house.groupby('bedrooms',as_index=False).price.mean()
average_price_per_room_num
```

draw bar graph that describe average relation between price and number of room

```python
sns.barplot(x="bedrooms",y="price",data=average_price_per_room_num)
```


draw line plot that describe average relation between price and number of room

```python
sns.lineplot(x="bedrooms",y="price",data=average_price_per_room_num)
```

draw lm plot that describe average relation between price and number of room

```python

sns.lmplot(x="bedrooms",y="price",data=average_price_per_room_num)
```
