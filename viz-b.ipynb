```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

import data data from csv file and show it

```python
df_life_expectancy = pd.read_csv('./life_expectancy_data.csv')
df_life_expectancy
```

calculate avg of life expectancy for each country

```python
average_life_expectancy_country=df_life_expectancy.groupby('Country',as_index=False).life_expectancy.mean().sort_values('life_expectancy', ascending=False).head(5)
average_life_expectancy_country
```

draw bar graph that describe average life expectancy of top five country

```python
sns.barplot(x="Country",y="life_expectancy",data=average_life_expectancy_country)
```

calculate releation life expectancy with Alcohol percentage

```python
releation_life_expectancy_with_Alcohol_percentage=df_life_expectancy[df_life_expectancy.Country=='Sweden'].groupby('life_expectancy',as_index=False).Alcohol.mean().sort_values('Alcohol', ascending=False)
releation_life_expectancy_with_Alcohol_percentage
```

draw line graph that describe releation between life expectancy and Alcohol percentage

```python
sns.lineplot(x="life_expectancy",y="Alcohol",data=releation_life_expectancy_with_Alcohol_percentage)
```
